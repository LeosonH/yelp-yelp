## Example 1: Hierarchical Clustering

```python
# Importing necessary packages
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as hr
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
dataset = load_iris()
iris_df = pd.DataFrame(dataset.data, columns =['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
iris_df.head()
```

```python
# complete dendrogram
# "complete": Performs complete/max/farthest point linkage clustering, see documentation https://docs.scipy.org/doc/scipy-0.14.0/reference/cluster.hierarchy.html
h_cluster = hr.complete(iris_df)
plt.figure(figsize = (10,8))
hr.set_link_color_palette(['purple','teal', 'maroon', 'orange'])
hr.dendrogram(h_cluster, labels = iris_df.index)
plt.title("Hierarchichal Dendrogram")
```

```python
# truncated
h_cluster = hr.complete(iris_df)
plt.figure(figsize = (10,8))
hr.set_link_color_palette(['purple','teal', 'maroon', 'orange'])
hr.dendrogram(h_cluster, truncate_mode='lastp', p=4, labels = iris_df.index)
plt.title("Hierarchichal Dendrogram (Truncated clusters = 4)")
```

```python
# Scaling, although not really necessary for the iris dataset since all variables are technically in the same scale
iris_scaled = StandardScaler().fit_transform(iris_df.values)
```

```python
h_scaled = hr.complete(iris_scaled)
plt.figure(figsize = (10,8))
hr.dendrogram(h_scaled, truncate_mode='lastp', p=4, labels = iris_df.index)
plt.title("Hierarchichal Dendrogram (Truncated clusters = 4, Scaled)")
```
